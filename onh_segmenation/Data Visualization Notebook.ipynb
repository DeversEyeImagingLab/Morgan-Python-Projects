# Comparing Mean Nearest Bundle Distance (NBD) vs Depth

## H005

```python
nbd_filename = "H005/H005_nbd_data.csv"

nbd_data = np.loadtxt(nbd_filename, delimiter=",")

# First, get data
# [z depth, mean nbd, 25th percentile, 75th percentile]
nbd_depth_data = []

nbd_data_wo_inf = nbd_data[np.where(nbd_data[:,3] != np.inf)]
z_slices = np.unique(nbd_data_wo_inf[:,0])

for depth in z_slices:
    nbd_subset = nbd_data_wo_inf[np.where(nbd_data_wo_inf[:,0] == depth)]
    nbd_depth_data.append([depth, 
                          np.mean(nbd_subset[:,3]), 
                          np.percentile(nbd_subset[:,3], 25), 
                          np.percentile(nbd_subset[:,3], 75)])

nbd_depth_data = np.asarray(nbd_depth_data)

# fit a line (we want to see if slope is >0, =0, or <0)
#array_x = np.vstack([bw_depth_data[:,0], np.ones(len(bw_depth_data[:,0]))]).T
#linear_m, linear_c = np.linalg.lstsq(array_x, bw_depth_data[:,1], rcond=None)[0]

fig, ax = plt.subplots()
ax.plot(nbd_depth_data[:,0], nbd_depth_data[:,1], '-')
ax.fill_between(nbd_depth_data[:,0], nbd_depth_data[:,2], nbd_depth_data[:,3], alpha=0.25)
#ax.plot(bw_depth_data[:,0], linear_m * bw_depth_data[:,0] + linear_c,'r-')
```

# Comparing Mean Pore Diameter vs Depth

## H005

```python
pore_diameter_filename = "H005_pore_diameter_data.csv"

# Load in data
# [Bundle number, slice x (depth), pore diameter]
pore_diameter_data = np.loadtxt(pore_diameter_filename, delimiter=",")

# First, get data
# [z depth, mean pd, 25th percentile, 75th percentile]
pd_depth_data = []

# Remove outliers
outlier_cutoff = np.percentile(pore_diameter_data[:,2], 95)
pd_data_no_outliers = pore_diameter_data[np.where(pore_diameter_data[:,2] <= outlier_cutoff)]

# Get all z slices
z_slices = np.unique(pd_data_no_outliers[:,1])

for depth in z_slices:
    pd_subset = pd_data_no_outliers[np.where(pd_data_no_outliers[:,1] == depth)]
    pd_depth_data.append([depth, 
                          np.mean(pd_subset[:,2]), 
                          np.percentile(pd_subset[:,2], 25), 
                          np.percentile(pd_subset[:,2], 75)])

pd_depth_data = np.asarray(pd_depth_data)

# fit a line (we want to see if slope is >0, =0, or <0)
array_x = np.vstack([pd_depth_data[:,0], np.ones(len(pd_depth_data[:,0]))]).T
linear_m, linear_c = np.linalg.lstsq(array_x, pd_depth_data[:,1], rcond=None)[0]

fig, ax = plt.subplots()
ax.plot(pd_depth_data[:,0], pd_depth_data[:,1], '-')
ax.fill_between(pd_depth_data[:,0], pd_depth_data[:,2], pd_depth_data[:,3], alpha=0.25)
ax.plot(pd_depth_data[:,0], linear_m * pd_depth_data[:,0] + linear_c,'r-')
```
