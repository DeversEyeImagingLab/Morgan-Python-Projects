```python
import skimage.io as skio
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
raw_data = skio.imread("combined_reflectance.tif", plugin="tifffile")
plt.imshow(raw_data[184,:,:])
```

```python
# Data info: 0    - no bundle
#            1-64 - bundle of group number x 

object_identities_data = np.load("combined_reflectance_Object Identities.npy")
plt.imshow(object_identities_data[184,:,:,0])
```

```python
# Data Organization
# [group_number, slice_x, cog_y, cog_z]

all_cog_data = []
```

```python
for slice_x in range(0, object_identities_data.shape[0]):
    
    # Which bundles are present on this slice
    bundles_on_slice = np.unique(object_identities_data[slice_x, :, :, 0])
    bundles_on_slice = np.delete(bundles_on_slice, 0)

    # Get one group
    for group_i in bundles_on_slice:    
        temp = np.zeros((object_identities_data.shape[1], object_identities_data.shape[2]))
        for yi in range(0, object_identities_data.shape[1]):
            for zi in range(0, object_identities_data.shape[2]):
                if object_identities_data[slice_x, yi, zi, 0] == group_i:
                    temp[yi, zi] = 1
    
        # Automatically find clusters
        labeled_array, num_features = ndimage.label(temp)
    
        # Find COG for each cluster
        for subgroup_i in range(1, num_features+1):
            y_locations = []
            z_locations = []
    
            # Average Y and Z coordinates
            for yi in range(0, labeled_array.shape[0]):
                for zi in range(0, labeled_array.shape[1]):
                    if labeled_array[yi,zi] == subgroup_i:
                        y_locations.append(yi)
                        z_locations.append(zi)
            
            # Add COGs with slice and group to data
            all_cog_data.append([group_i, slice_x, np.mean(y_locations), np.mean(z_locations)])
               
```

```python
# Convert the COG data to a numpy usable format
all_cog_data = np.asarray(all_cog_data)
```

```python
# Save cog data to easily reload later
all_cog_data.tofile("axon_bundle_cog_data.bin")
```
