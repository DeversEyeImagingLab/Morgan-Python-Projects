```python
import skimage.io as skio
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
object_identities_data = np.load("combined_reflectance_Object Identities.npy")
plt.imshow(object_identities_data[200,:,:,0])
```

```python
# Remove all group assignments
for xi in range(0,object_identities_data.shape[0]):
    for yi in range(0,object_identities_data.shape[1]):
        for zi in range(0,object_identities_data.shape[2]):
            if object_identities_data[xi,yi,zi,0] != 0:
                object_identities_data[xi,yi,zi,0] = 255
```

```python
plt.imshow(object_identities_data[200,:,:,0])
```

```python
# Save data to tiff
skio.imsave("bundles_no_groups.tif", object_identities_data, plugin="tifffile")
```
