```python
import skimage.io as skio
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

Input labels (as tif) filename:

```python
labels_data_filename = "H001/H001_OS_bundles.tif"
```

Output Filenames:

```python
cog_data_output_file_prefix = "H001_OS_cog_data"
tortuosity_data_output_file_prefix = "H001_OS_tortuosity_data"
```

```python
labels_data = skio.imread(labels_data_filename, plugin="tifffile")
plt.imshow(labels_data[200,:,:])
```

```python
# Data Organization
# [group_number, slice_x, cog_y, cog_z]

all_cog_data = []
```

```python
for slice_x in range(0, labels_data.shape[0]):
    
    # Which bundles are present on this slice
    bundles_on_slice = np.unique(labels_data[slice_x, :, :])
    bundles_on_slice = np.delete(bundles_on_slice, 0)

    # Get one group
    for group_i in bundles_on_slice:
        
        # Clear temp data
        y_locations = []
        z_locations = []
        
        # Average Y and Z coordinates
        for yi in range(0, labels_data.shape[1]):
            for zi in range(0, labels_data.shape[2]):
                if labels_data[slice_x, yi, zi] == group_i:
                    y_locations.append(yi)
                    z_locations.append(zi)
    
        # Add COGs with slice and group to data
        all_cog_data.append([group_i, slice_x, np.mean(y_locations), np.mean(z_locations)])
               
```

```python
# Convert the COG data to a numpy usable format
all_cog_data = np.asarray(all_cog_data)
```

```python
# Save cog data to easily reload later
#all_cog_data.tofile((cog_data_output_file_prefix + ".bin"))
#all_cog_data.tofile((cog_data_output_file_prefix + ".txt"), sep=',')
np.savetxt((cog_data_output_file_prefix + ".csv"), all_cog_data, delimiter=",")
```

```python
# Filter by label
group_data = all_cog_data[np.where(all_cog_data[:,0] == 1)]
group_data_b = all_cog_data[np.where(all_cog_data[:,0] == 2)]
group_data_c = all_cog_data[np.where(all_cog_data[:,0] == 4)]

# Plot Y by X view
plt.scatter(group_data[:,3], -group_data[:,1], s=20, c='blue')
plt.scatter(group_data_b[:,3], -group_data_b[:,1], s=20, c='red')
plt.scatter(group_data_c[:,3], -group_data_c[:,1], s=20, c='green')
plt.show()
```

```python
# Plot Y by X view
plt.scatter(all_cog_data[:,3], -all_cog_data[:,1], s=20, c='blue')
plt.show()
```

```python
# For each bundle, we should find L and L'
# [Bundle number, L, L', tortuosity]
tortuosity_data = []
```

```python
# Loop over each bundle
bundles = np.unique(all_cog_data[:,0])
for bundle_number in bundles:
    
    # Get data for only this bundle
    this_bundle_data = all_cog_data[np.where(all_cog_data[:,0] == bundle_number)]
    
    # Find L
    l_val = np.max(this_bundle_data[:,1]) - np.min(this_bundle_data[:,1])
    
    # Find L' 
    l_prime_list = []
    for i in range(0, len(this_bundle_data) - 1):
        # Find Li' using Pythagorean theorem for 3 dimensions
        l_prime_i = np.sqrt((this_bundle_data[i,1] - this_bundle_data[i+1,1]) **2 +
            (this_bundle_data[i,2] - this_bundle_data[i+1,2]) **2 +
            (this_bundle_data[i,3] - this_bundle_data[i+1,3]) **2)
        
        l_prime_list.append(l_prime_i)
    
    l_prime = sum(l_prime_list)
    
    # Calculate tortuosity
    t_val = l_prime / l_val
    
    # Add to tortuosity data
    tortuosity_data.append([bundle_number, l_val, l_prime, t_val])
```

```python
# Convert the tortuosity_data to a numpy usable format
tortuosity_data = np.asarray(tortuosity_data)
```

```python
# Save tortuosity data to easily reload later
#tortuosity_data.tofile("manual_label_tortuosity_data.bin")
#tortuosity_data.tofile("manual_label_tortuosity_data.txt", sep=',')
np.savetxt((tortuosity_data_output_file_prefix + ".csv"), tortuosity_data, delimiter=",")
```
